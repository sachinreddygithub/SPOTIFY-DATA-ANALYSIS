# # **most popular songs**

```
most_popular=spoti.query('popularity>90', inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```
spoti.set_index('release_date',inplace=True)
spoti.index=pd.to_datetime(spoti.index)
spoti.head()
```

```
spoti.duration_ms.head
```

# # **Dropping the unwanted col and applying .corr() fun**

```
corr_df=spoti.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno',linewidth=1, linecolor='Black')
heatmap.set_title('Correlation HeatMap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

# # **sampalise the .04 out of tot and create regre graphs** 

```
sample_df=spoti.sample(int(0.004*len(spoti)))
```

```
print(len(sample_df))
```

```
plt.figure(figsize=(10,6))
sns.regplot(data =sample_df, y='loudness',x='energy', color='c').set(title= 'Loudness vs Energy Correlation')
```

```
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x='acousticness', y='popularity', color='b').set(title='popularity vs acousticness correlation')
```

# # **Releasdate to date**

```
spoti['dates']=spoti.index.get_level_values('release_date')
spoti.dates=pd.to_datetime(spoti.dates)
years=spoti.dates.dt.year
```

# # **Distribution plot**

```
sns.displot(years,discrete=True,aspect=2,height=5, kind='hist').set(title='Number of songs per year')
```

# # **ratio of songs per yr**

```
total_dr=spoti.duration_ms
fig_dims =(18,7)
fig, ax= plt.subplots(figsize = fig_dims)
fig=sns.barplot(x=years, y=total_dr,ax=ax, errwidth=False).set(title='Year vs Duration_ms')
plt.xticks(rotation=90)
```
